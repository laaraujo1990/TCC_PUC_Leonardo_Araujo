```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Carrega os dois dataframes
raw_data_enem = pd.read_csv('./MICRODADOS_ENEM.zip', compression='zip' , sep=';', encoding="latin_1")
raw_data_idd = pd.read_csv('./MICRODADOS_IDD.zip', compression='zip' , sep=';', encoding="latin_1")

# Combina os dois dataframes com base na coluna "NU_INSCRICAO"
raw_data = pd.merge(raw_data_enem, raw_data_idd, on='NU_INSCRICAO')
raw_data.head()
```

```python
raw_data.iloc[1,:]
```

```python
#excluindo dados de gabarito e respostas
raw_data.drop(labels=['TX_GABARITO_CN','TX_GABARITO_CH','TX_GABARITO_LC','TX_GABARITO_MT',
                      'TX_RESPOSTAS_CN','TX_RESPOSTAS_CH','TX_RESPOSTAS_LC','TX_RESPOSTAS_MT',
                      'CO_PROVA_CN','CO_PROVA_CH','CO_PROVA_LC','CO_PROVA_MT'], 
              inplace=True, axis=1)

#excluido dados municipio e uf de residencia
raw_data.drop(labels=['CO_MUNICIPIO_RESIDENCIA',
                      'CO_MUNICIPIO_NASCIMENTO','NO_MUNICIPIO_NASCIMENTO',
                      'CO_UF_NASCIMENTO','SG_UF_NASCIMENTO'], 
              inplace=True, axis=1)

#excluindo dados de atendimento especializado
raw_data.drop(labels=[
'IN_BAIXA_VISAO',
'IN_CEGUEIRA',
'IN_SURDEZ',
'IN_DEFICIENCIA_AUDITIVA',
'IN_SURDO_CEGUEIRA',
'IN_DEFICIENCIA_FISICA',
'IN_DEFICIENCIA_MENTAL',
'IN_DEFICIT_ATENCAO',
'IN_DISLEXIA',
'IN_DISCALCULIA',
'IN_AUTISMO',
'IN_VISAO_MONOCULAR',
'IN_OUTRA_DEF',
    
'IN_SEM_RECURSO',
'IN_BRAILLE',
'IN_AMPLIADA_24',
'IN_AMPLIADA_18',
'IN_LEDOR',
'IN_ACESSO',
'IN_TRANSCRICAO',
'IN_LIBRAS',
'IN_TEMPO_ADICIONAL',
'IN_LEITURA_LABIAL',
'IN_MESA_CADEIRA_RODAS',
'IN_MESA_CADEIRA_SEPARADA',
'IN_APOIO_PERNA',
'IN_GUIA_INTERPRETE',
'IN_COMPUTADOR',
'IN_CADEIRA_ESPECIAL',
'IN_CADEIRA_CANHOTO',
'IN_CADEIRA_ACOLCHOADA',
'IN_PROVA_DEITADO',
'IN_MOBILIARIO_OBESO',
'IN_LAMINA_OVERLAY',
'IN_PROTETOR_AURICULAR',
'IN_MEDIDOR_GLICOSE',
'IN_MAQUINA_BRAILE',
'IN_SOROBAN',
'IN_MARCA_PASSO',
'IN_SONDA',
'IN_MEDICAMENTOS',
'IN_SALA_INDIVIDUAL',
'IN_SALA_ESPECIAL',
'IN_SALA_ACOMPANHANTE',
'IN_MOBILIARIO_ESPECIFICO',
'IN_MATERIAL_ESPECIFICO',
'IN_NOME_SOCIAL'], inplace=True, axis=1)


raw_data.drop(labels=['CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA','CO_UF_PROVA','SG_UF_PROVA'],
              inplace=True, axis=1)
```
